### Paso 5: Obtener los seguidores (followers)

```python
# Obtener el URL de tu cuenta
tu_cuenta = driver.find_element_by_xpath("//a[@class='gmFkV']").get_attribute('href')
```

```python
# Acceder a tu cuenta
driver.get(tu_cuenta)
```

```python
# Obtener el número de seguidores - [0] porque primero es 'seguidores'
num_followers = int(driver.find_elements_by_xpath("//ul//li//a//div//span[@class='g47SY ']")[0].text.replace(',', ''))
num_followers
```

```python
# Ingresar a tus seguidores - [0] porque primero es 'seguidores'
driver.find_elements_by_xpath("//ul//li//a//div//span[@class='g47SY ']")[0].click()
```

```python
# Mantenerse en el pop-up
pop_up_window = WebDriverWait(driver, 2).until(EC.element_to_be_clickable((By.XPATH, "//div[@class='isgrP']")))

# Scrollear todos los seguidores
b=0
while num_followers > b:
    driver.execute_script('arguments[0].scrollTop = arguments[0].scrollTop + arguments[0].scrollTop + arguments[0].offsetHeight;', pop_up_window)
    b=len(driver.find_elements_by_xpath("//a[@class='notranslate _0imsa ']"))
    time.sleep(1)
```

```python
# Obtener la cuenta de cada uno de los seguidores
obtener_followers = driver.find_elements_by_xpath("//a[@class='notranslate _0imsa ']")

followers = []
for i in obtener_followers:
    follower = i.get_attribute('title')
    followers.append(follower) 
followers
```

### Paso 6: Obtener los seguidos (following)

```python
# Acceder a tu cuenta
driver.get(tu_cuenta)

# Obtener el número de seguidos - [1] porque segundo es 'seguidos'
num_following = int(driver.find_elements_by_xpath("//ul//li//a//div//span[@class='g47SY ']")[1].text)

# Ingresar a tus seguidos - [1] porque segundo es 'seguidos'
driver.find_elements_by_xpath("//ul//li//a//div//span[@class='g47SY ']")[1].click()

# Mantenerse en el pop-up
pop_up_window = WebDriverWait(driver, 2).until(EC.element_to_be_clickable((By.XPATH, "//div[@class='isgrP']")))

# Scrollear todos los seguidos
b=0
while num_following > b:
    driver.execute_script('arguments[0].scrollTop = arguments[0].scrollTop + arguments[0].scrollTop + arguments[0].offsetHeight;', pop_up_window)
    b=len(driver.find_elements_by_xpath("//a[@class='notranslate _0imsa ']"))
    time.sleep(1)

# Obtener la cuenta de cada uno de los seguidos
obtener_following = driver.find_elements_by_xpath("//a[@class='notranslate _0imsa ']")
followings = []
for i in obtener_following:
    following = i.get_attribute('title')
    followings.append(following) 
followings
```

### Paso 7: Ver las cuentas que seguimos, pero no nos siguen

```python
dejar_de_seguir = list(set(followings) - set(followers))
dejar_de_seguir
```

### Paso 8: No dejar de seguir a las cuentas que tienen más de cierta cantidad de seguidores (deben ser páginas o personas reconocidas)

```python
no_dejar_de_seguir = []
for cuenta in tqdm(dejar_de_seguir):
    driver.get('https://www.instagram.com/' + cuenta)
    num_followers = driver.find_elements_by_xpath("//ul//li//a//div//span[@class='g47SY ']")[0].text.replace(',', '')
    
    if 'k' in num_followers or 'm' in num_followers:
        # Las cuentas cuyos seguidores tienen k (miles) o m (millones) quiere decir que su num_followers >= 10000
        no_dejar_de_seguir.append(cuenta)
    elif int(num_followers) >= 8000:
        # Si una cuenta tiene más 8000 seguidores y no nos sigue, no se va a dejar de seguir porque se asume que es una cuenta reconocida.
        # Este número puede variar de acuerdo al resultado que se desee
        no_dejar_de_seguir.append(cuenta)

dejar_de_seguir = [x for x in dejar_de_seguir if x not in no_dejar_de_seguir]
dejar_de_seguir 
```
